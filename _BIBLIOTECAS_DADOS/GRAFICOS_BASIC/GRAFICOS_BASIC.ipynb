# Como montar gráficos?
**Usando matplotlib.pyplot, pandas, numpy, os e math**

***IMPORTAÇÕES***

```python
%matplotlib inline
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

import os
import math
```

***DADOS PARA MONTAR OS GRÁFICOS***

```python
x = []
graus = []
for i in range(0,25):
    x.append(i)
    graus.append(i*15)

radiano = [0]*len(graus)
for g in range(len(graus)):
    radiano[g] = math.radians(graus[g])

yy = []
for n in range(len(radiano)):
    yy.append(3*radiano[n])

A = []
B = []
N = []
for n in range(0,10):
    A.append(81*n)
    B.append(n**3)
    N.append(n)
    
dados = pd.read_excel("DADOS.xlsx")
coluna3 = np.random.choice(["c", "d","e"], size= len(dados))
dados["TIPO"] = coluna3
DADOS = dados.TIPO.value_counts()
dados.to_csv("dados_salvos.csv")
```

> MARKERS
* ***-*** = point marker
* ***,*** = pixel marker
* ***o*** = circle marker
* ***v*** = triangle_down marker
* ***^*** = triangle_up marker
* ***<*** = triangle_left marker
* ***>*** = triangle_right marker
* ***1*** = tri_down marker
* ***2*** = tri_up marker
* ***3*** = tri_left marker
* ***4*** = tri_right marker
* ***s*** = square marker
* ***p*** = pentagon marker
* ***' * '*** = star marker
* ***h*** = hexagon1 marker
* ***H*** = hexagon2 marker
* ***+*** = plus marker
* ***x*** = x marker
* ***D*** = diamond marker
* ***d*** = thin_diamond marker
* ***|*** = vline marker
* ***_*** = hline marker

> LINE STYLES
* ***-*** = solid line
* ***--*** = dashed line
* ***-.*** = dashed-dot line
* ***:*** = dotted line

> CORES
* ***b*** = blue
* ***g*** = green
* ***r*** = red
* ***c*** = cyan
* ***m*** = magenta
* ***y*** = yellow
* ***k*** = black
* ***w*** = white

**GRÁFICOS LINEARES E GERAIS**

```python
plt.plot(N, A, "r-o", label="A")
plt.plot(N, B, "k--o", label="B")

plt.autoscale(enable=True, axis='both', tight=False)

#plt.arrow(x0,y0,x,y, alpha=1, color=x)
#linestyle={'solid', 'dashed', 'dashdot', 'dotted'}
plt.arrow(0, 0, 8, 500, alpha=0.5, linestyle='dashdot')
#plt.vlines(x,ymin,ymax,...)
plt.vlines(8, 0, 500, alpha=0.5, linestyle='dashdot')

plt.grid(True)

plt.box(on=False)

#fontsize : int or float or {'xx-small', 'x-small', 'small', 'medium', 'large', 'x-large', 'xx-large'}
#loc='best' 'upper right' 'upper left' 'lower left' 'lower right'
#    'right' 'center left' 'center right' 'lower center' 'upper center' 'center' (or number)
# markerfirst=False coloca o texto antes
# shadow=True deixa a caixa mais sólida, com sombra
# facecolor, edgecolor, title, title_fontsize

plt.legend(loc='best', fontsize='large', markerfirst=False, shadow=True, edgecolor='k', title="LEGENDA", title_fontsize='large')

#plt.text(x,y,'texto', fontsize=10, 
# if:: bbox=dict(**)) 
# >> adiciona uma caixa em volta do texto
# ** > facecolor, alpha
plt.text(5, 30, 'curva preta', fontsize=13, bbox=dict(facecolor='w', alpha=0.5) )
plt.text(2, 430, 'curva vermelha', fontsize=13, bbox=dict(facecolor='w', alpha=0.5) )

#deixa de forma hashurada
plt.minorticks_on()

plt.xlabel("EIXO X");
plt.ylabel("EIXO Y");

plt.title("TITULO DO GRÁFICO\n[Nunca se esqueça de por título!]", fontsize=15, loc='left', pad=12);

plt.xlim(0, 9)
plt.ylim(0, 800)

plt.show()
```

**GRÁFICOS POLARES**

**GRÁFICOS DE BARRA**

```python
#plt.bar(x, height, width=0.8, bottom=None (início da barra em y), align='center')
plt.bar(N, A, width=0.5, bottom=None, align='edge', color='c', edgecolor='k', log=False, alpha=0.90)
plt.bar(N, B, width=0.5, bottom=None, align='center', color='g', edgecolor='k', log=False, alpha=0.90)
plt.show()
```
